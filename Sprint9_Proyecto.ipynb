 <font size="4"> **¬øCu√°ntas personas lo usan cada d√≠a, semana y mes?** </font>

```python
# Para responder esta pregunta, debemos realizar los calculos con las formulas correctas para descubrir la cantidad de personas que usan la aplicaci√≥n.

dau_total = visits_df.groupby('session_date').agg({'uid': 'nunique'}).mean()
wau_total = visits_df.groupby(['session_year', 'session_week']).agg({'uid': 'nunique'}).mean()
mau_total = (visits_df.groupby(['session_year', 'session_month']).agg({'uid': 'nunique'}).mean())

print('N√∫mero de usuarios activos diarios (unicos): ', dau_total)
print()
print('N√∫mero de usuarios activos semanales: ', wau_total)
print()
print('N√∫mero de usuarios activos mensuales: ', mau_total)
```

Gracias a los calculos que realizamos descubrimos que 908 fueron los usuarios que utilizan el servicio diaramente, 5,716 la usan semanalmente y 23228 la usan mensualmente. Se podria considerar una buena cantidad de usuarios, vamos a seguir analizando.

<div class="alert alert-block alert-success"> <b>Comentario del revisor</b> <a class="tocSkip"></a><br>
<b>√âxito</b> - Muy bien al calcular los usuarios activos diarios, semanales y mensuales utilizando agrupaciones y agregaciones adecuadas. Presentar los promedios brinda un panorama claro del uso del servicio en diferentes per√≠odos.

</div>

 <font size="4"> **¬øCu√°ntas sesiones hay por d√≠a? (Un usuario puede tener m√°s de una sesi√≥n).** </font>

```python
# Para descrubrir esta informaci√≥n, primero necesitamos agrupar los datos con las columnas "year" y "date" para encontrar el numero de usuarios unicos del servicio.

sessions_per_user = visits_df.groupby(['session_year', 'session_date']).agg(
    {'uid': ['count', 'nunique']}
)

# Vamos a cambiar los nombres de las columnas para que se entiendan mejor 

sessions_per_user.columns = ['n_sessions', 'n_users']

# Aplicaremos la formula para encontrar la cantidad de sesiones por usuario diario

sessions_per_user['sessions_per_user'] = (
    sessions_per_user['n_sessions'] / sessions_per_user['n_users']
)
sessions_per_user= sessions_per_user.reset_index()
print(sessions_per_user)
```

```python
# Una vez teniendo los resultados, vamos a crear un histograma para poder observar a detalle la distribuci√≥n del numero de sesiones de usuarios que se tiene para cada dia.

sessions_per_user['sessions_per_user'].plot(kind= 'hist',
                                        bins= 100,
                                       title= 'N√∫mero de sesiones por usuario',
                                       figsize= (12, 8),
                                       color= 'blue'
                                       )
plt.xlabel('N√∫mero de sesiones diarias')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Ahora realizamos el mismo procedimiento pero para graficar las sesiones de usuarios para cada mes.

sessions_per_user.plot(x = 'session_date',
                  y= 'sessions_per_user',
                  kind= 'line',
                  figsize= [12,8],
                  fontsize= 12,
                  rot= 100,
                  color= 'green',
                  legend=False     
                       )
plt.title('Numero de sesiones por usuarios mensuales', fontsize=15)
plt.xlabel('Mes', fontsize=15)
plt.ylabel('Cantidad promedio de sesiones', fontsize=15)


plt.show()
```

Como podemos observar en nuestro histograma, contamos con un numero basico y normal de usuarios por dia que en este caso el promedio es 1. Y en el grafico mensual de usuarios, podemos observar que el mes que m√°s destaca es el de Diciembre a comparaci√≥n de los demas meses.

<div class="alert alert-block alert-success"> <b>Comentario del revisor</b> <a class="tocSkip"></a><br>
<b>√âxito</b> - Buen trabajo al calcular el promedio de sesiones por usuario y representarlo visualmente mediante un histograma y una gr√°fica temporal. Esto permite identificar patrones diarios y mensuales, destacando eficientemente per√≠odos clave como diciembre.

</div>

 <font size="4"> **¬øCu√°l es la duraci√≥n de cada sesi√≥n?** </font>

```python
# Primero necesitamos crear una nueva columna que se llamara "session_duration_sec" y aplicaremos formulas para encontrar la duraci√≥n de cada sesi√≥n.
visits_df['session_duration_sec'] = (visits_df['end_ts'] - visits_df['start_ts']).dt.seconds
print(visits_df['session_duration_sec'].mean())
```

```python
# Teniendo el resultado, vamos a crear un histograma para poder visualizar mejor los datos.
visits_df['session_duration_sec'].hist(bins=100)
```

```python
# Notamos que los datos no son normales, asi que vamos a calcular la moda.
asl= visits_df['session_duration_sec'].mode()
print(asl)
```

```python
# Teniendo la moda, utilizamos el metodo describe() para poder tener una vista m√°s clara y completa de como se comportan los datos en nuestra columna
visits_df['session_duration_sec'].describe()
```

Como pudimos observar, al principio el histograma nos arrojo valores atipicos, por eso se decidio calcular la moda. El describe nos dice que el numero maximo es de 84480 segundos, lo cual en minutos son 1408, lo cual viene siendo practicamente un dia. En cambio, la moda nos arrojo que la duraci√≥n promedio de un usuario es de 60 segundos, lo que viene siendo 1 minuto practicamente. Ese resultado ya es m√°s normal.

<div class="alert alert-block alert-success"> <b>Comentario del revisor</b> <a class="tocSkip"></a><br>
<b>√âxito</b> - Excelente enfoque al calcular la duraci√≥n de las sesiones y analizar los valores at√≠picos con histogramas y estad√≠sticas descriptivas. Usar la moda como medida representativa fue una decisi√≥n acertada para interpretar datos con sesgo.

</div>

 <font size="4"> **¬øCon qu√© frecuencia los usuarios regresan?** </font>

```python
# Para calcular la frecuencia, necesitamos utilizar las formulas con "sticky"

sticky_wau= dau_total/wau_total * 100
print(sticky_wau)
sticky_mau= dau_total/mau_total * 100
print(sticky_mau)
```

Como podemos observar:
- sticky_wau: Nos dio como resultado 15.88%, que vienen siento los usuarios que usan el servicio en una semana tambien la usan diariamente.

- sticky_mau: Nos dio como resultado 3.90%, que vienen siendo los usuarios que usan el servicio en un mes y tambien la usan diariamente.

Contamos con un compromiso adecuado de es del 16%, el cual es optimo y sugiere que existen usuarios que se encuentran utilizando el servicio diariamente.

```python
# Teniendo esta informaci√≥n, ya podemos buscar la primera sesi√≥n de cada usuario
first_session_dates = visits_df.groupby('uid')['month_session'].min().reset_index()

# Cambiamos el nombre de la columna
first_session_dates.columns = ['uid', 'first_session_month']

# Comprobamos si se aplicaron correctamente los cambios
first_session_dates.head()
```

```python
# Ya que tenemos esta columna, podemos unirla al primer DataFrame con el metodo de "merge"
visits_log_us_ = visits_df.merge(first_session_dates, on= 'uid')
visits_log_us_.head(5)
```

```python
# Teniendo esta informaci√≥n, ya podemos buscar la fecha para la primera orden que realizo cada usuario.
first_buy_dates = orders_df.groupby('uid')['order_month'].min().reset_index()

# De nuevo, actualizamos el nombre de la columna 
first_buy_dates.columns = ['uid', 'first_buy_month']

# Comprobamos que se hayan realizado correctamente los cambios
first_buy_dates.head()
```

```python
# Unimos esta columna a nuestro segundo DataFrame con el metodo de "merge"
orders_log_us_ = orders_df.merge(first_buy_dates, on= 'uid')
orders_log_us_.head(5)
```

```python
# Juntamos los dos DataFrames "visits_log_us" y "orders_log_us" en uno solo
visits_orders = visits_log_us_.merge(orders_log_us_, on= 'uid')
visits_orders.head(5)
```

```python
# En este nuevo DataFrame, calculamos los dias transcurridos para cuando un usuario se convierte en cliente
visits_orders['convertion_time_days'] = (visits_orders['first_buy_month'] - visits_orders['first_session_month']).dt.days
visits_orders.head(5)
```

```python
# Empezamos a categorizar el tiempo de conversi√≥n

# Definimos los intervalos que debe de tener
bins = [-1, 0, 1, 7, 30, 60, 90, 120, 150, 180, 210, 240, 270, 300, 330, 360]
labels = ['Conversion 0d', 'Conversion 1d', 'Conversion 1w', 'Conversion 1m', 'Conversion 2m', 'Conversion 3m', 'Conversion 4m', 'Conversion 5m', 'Conversion 6m', 'Conversion 7m', 'Conversion 8m', 'Conversion 9m', 'Conversion 10m', 'Conversion 11m', 'Conversion 12m']
visits_orders['conversion_category'] = pd.cut(visits_orders['convertion_time_days'], bins=bins, labels=labels)

# Imprimimos las filas
visits_orders.sample(5)
```

```python
# Ya que tenemos nuestra informaci√≥n filtrada, podemos crear una tabla dinamica para saber la cantidad de pedidos que hicieron los usuarios por Cohorte
first_session_cohort = visits_orders.pivot_table(index= 'first_session_month',
                                       columns= 'conversion_category',
                                       values= 'uid',
                                       aggfunc= 'nunique')
```

```python
# Una vez obtenida nuestra tabla dinamica, podemos graficar un mapa de calor para que nos muestre la informaci√≥n m√°s clara y precisa.
plt.figure(figsize=(18, 9))

sns.heatmap(first_session_cohort, annot=True, fmt='g', cmap="crest", linewidths= 1, linecolor='black')

plt.title('Cantidad de usuarios por primer pedido', fontsize= 16)
plt.xlabel('Categor√≠a por conversi√≥n', fontsize= 14)


plt.show()
```

```python
# Ahora que los Cohores se definen por el periodo de tiempo de conversi√≥n vamos a crear una tabla din√°mica para calcular la cantidad de pedidos que se hicieron
convertion_time_cohort = visits_orders.pivot_table(index= 'conversion_category',
                             columns= 'source_id',
                             values= 'uid',
                             aggfunc= 'nunique')
```

<font size="4"> **¬øCu√°nto dinero traen? (LTV)** </font>

```python
# Creamos el Cohorte para buscar dinero que traen los diferentes usuarios
cohort_sizes = orders_log_us_.groupby('first_buy_month').agg({'uid': 'nunique'}).reset_index()

# Actualizamos el nombre de las columnas
cohort_sizes.columns = ['first_buy_month', 'n_buyers']

# Imprimimos resultados
cohort_sizes.head()
```

```python
# Ahora, agrupamos la informaci√≥n la orden del mes
cohorts = orders_log_us_.groupby(['first_buy_month','order_month']).agg({'revenue': 'sum'}).reset_index()
cohorts.head()
```

```python
# Creamos un reporte en donde unimos los dos cohortes en la columna de "first_buy_month"
report = pd.merge(cohort_sizes, cohorts, on='first_buy_month')
report.head()
```

```python
# Una vez obtenido nuestro reporte, ya podemos calcular el LTV
report['age'] = (
    report['order_month'] - report['first_buy_month']
) / np.timedelta64(1, 'M')
report['age'] = report['age'].round().astype('int')

report['ltv'] = report['revenue'] / report['n_buyers']

report.head()
```

```python
# Creamos una tabla dinamica con los resultados del reporte
output = report.pivot_table(
    index='first_buy_month', columns='age', values='ltv', aggfunc='mean'
)

output.fillna('')
```

```python
# Una vez obtenida la tabla, podemos graficarlos con un mapa de calor para tener una informaci√≥n m√°s clara y precisa
plt.figure(figsize=(18, 9))

sns.heatmap(output, annot=True, fmt='.2f', cmap="crest",  linewidths= 1, linecolor='black')

plt.title('LTV promedio de los clientes.', fontsize= 16)
plt.xlabel('Edad de los cohortes', fontsize= 14)


plt.show()
```

```python
# Por ultimo, realizamos la suma total de los ingresos obtenidos por los usuarios en los diferentes meses.
output.sum(axis=1).round()
```

Como podemos observar, podemos apreciar que se generan ingresos desde la edad del cohorte 0, lo cual nos indica que existe una gran cantidad de usuarios que realizar su primera compra el primer momento en que se registran en el servicio. Ademas, se mantiene un estandar de entre 6 y 8 a finales del a√±o 2017 e inicios del 2018 hasta llegar a Junio que hay un bajor de 3.
